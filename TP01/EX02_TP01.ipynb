II. Transformée de Fourier discrète par FFT d’un signal périodique
Reprendre le programme ExemplePreliminaire et modifier la fréquence du signal.
Esse exercicio é para entender o efeito do numero de potnos que nos usamos em nosso codigo

```python
# -*- coding: utf-8 -*-
# o algoritimo da FFT parte do principio que a gente observa pelo menos um periodo inteiro
# Ou seja é preciso que o tamanho da janela seja um multiplo do periodo
# Observamos que quando nao temos um numero inteiro de periodos, vemos uma reparticao do spectro
# Existe um link forte entre o numero de amostras e o periodo do sinal 
"""
Created on Sun Feb 23 14:49:13 2025

@author: guyadern
"""

import matplotlib.pyplot as plt
import numpy as np
#essa linha é so para quando a gente for chamar np.fft.fft eu irei escrever fft(como um using namespace em C)
from numpy.fft import fft, ifft, fftshift
```

```python
N=10; fo=0.1; Fe=1; Te=1/Fe
k=np.arange(0,N,1) # N nao esta incluso na lista, ela vai de 0 a N-1
t=Te*k
x_k=np.sin(2*np.pi*fo*t)

#% on crée un axe des temps 'continu'
tc=Te*np.arange(0,N,Te) # aqui a gente finge que esta no continuo para tracar o sinal
xc=np.sin(2*np.pi*fo*tc)
# Une façon de visualiser le signal et le signal échantillonné en tenant compte du temps en secondes
plt.figure()
plt.stem(t,x_k,label='Signal échantilonné');
plt.xlabel('Temps (ms)');plt.ylabel('Signal')
plt.legend()

# On calule la TF
X_f=fft(x_k);
# axe des fréquences réduites 
x_f=np.arange(0,1,1/N);
# axe des fréquences
freq=x_f*Fe;

X_fc=fftshift(X_f)
x_fc=np.arange(-1/2,1/2,1/N)
freqc=x_fc*Fe;

plt.figure()
plt.stem(freqc,np.abs(X_fc)/N);
plt.xlabel('Fréquence'); plt.ylabel('Module du spectre');
```
