```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.fft import fft, ifft, fftshift
```

```python
#Aqui é para os sinais 
import scipy.signal.windows as scw
```

```python
#N → Define o formato da janela no tempo, afetando o espalhamento no espectro de frequência.
# 1024 → Define a resolução da FFT. 
# Um valor grande aumenta a suavidade do espectro sem mudar sua informação fundamental
fe=1; N=64
k = np.arange(0, N)

rectangle=np.zeros(N)
rectangle[1:N-1]=1
ham = scw.hamming(N)
blk = scw.blackman(N)
plt.plot(k, rectangle, label = "Rectangle")
plt.plot(k, ham, label = "Hamming")
plt.plot(k, blk, label = "Blackman")
plt.legend()
```

```python
#FFT
Rect_f = fftshift(fft(rectangle))
Ham_f = fftshift(fft(ham))
Blk_f = fftshift(fft(blk))
x_f = np.arange(-1/2, 1/2, 1/N)

plt.figure()
plt.plot(x_f, 10*np.log10(np.abs(Rect_f)/len(rectangle)))


plt.figure()

# plt.stem(x_f, 10*np.log10(np.abs(Ham_f)/np.len(ham)))
```
