```python
import numpy as np
import matplotlib.pyplot as plt
```

PARTIE 1- IMPULSION

```python
#-Créer une impulsion unité en k=0 (longueur du signal N=16 points).
# Visualiser sur une même figure ce signal, ainsi que le module (commande abs) et la phase
# (commande angle) de sa TF (commande fft).

N=16 
k= np.arange(0,N,1)
x_k= np.zeros(len(k))
x_k[0]=1 #adicionando os 1

X_f=np.fft.fft(x_k)
module_X=np.abs(X_f)
argument_X=np.angle(X_f)

plt.figure()
plt.subplot(3, 1, 1) #plot 1
plt.stem(k, x_k, label="x(k)")
plt.xlabel("k")
plt.ylabel("x[k]")
plt.legend()

plt.subplot(3, 1, 2) #plot 2
plt.plot(k, module_X, label="|X(f)|")
plt.xlabel("Frequence (k)")
plt.ylabel("Amplitude")
plt.legend()

plt.subplot(3, 1, 3) #plot 3
plt.plot(k, argument_X, label="arg(X(f))")
plt.xlabel("Frequence(k)")
plt.ylabel("Phase")
plt.legend()
```

**Analyse des résultats :**  
On observe que la Transformée de Fourier d’une impulsion correspond à la théorie. Nous voyons que la transformée d’une impulsion unitaire centrée en 0 donne une magnitude constante (1) en fréquence et une phase nulle (sans retard)

```python
#-Créer une impulsion unité en k=+4 (longueur N=16 points). 
N=16 
k= np.arange(0,N,1)
x_k= np.zeros(len(k))
x_k[4]=1 #adicionando os 1

X_f=np.fft.fft(x_k)
module_X=np.abs(X_f)
argument_X=np.angle(X_f)

plt.figure()
plt.subplot(3, 1, 1) #plot 1
plt.stem(k, x_k, label="x(k)")
plt.xlabel("k")
plt.ylabel("x[k]")
plt.legend()

plt.subplot(3, 1, 2) #plot 2
plt.plot(k, module_X, label="|X(f)|")
plt.xlabel("Frequence (k)")
plt.ylabel("Amplitude")
plt.legend()

plt.subplot(3, 1, 3) #plot 3
plt.plot(k, argument_X, label="arg(X(f))")
plt.xlabel("Frequence(k)")
plt.ylabel("Phase")
plt.legend()
```

# Analyse des résultats

On observe que la Transformée de Fourier d’une impulsion décalée correspond à la théorie. Nous voyons que la transformée d’une impulsion unitaire décalée en \( +4 \) donne une magnitude constante (1) en fréquence, comme l’impulsion centrée à 0, mais la phase, par contre, présente un retard.  

On sait que la transformée d’un signal décalé est donnée par :  

\[
X_{\text{décalé}}(f) = X(f) \cdot e^{-j 2\pi f \cdot \text{retard}}
\]



```python
N=16 
k= np.arange(0,N,1)
x_k= np.zeros(len(k))
x_k[0:5]=1 #adicionando os 1

X_f=np.fft.fft(x_k)
module_X=np.abs(X_f)
argument_X=np.angle(X_f)

plt.figure()
plt.subplot(3, 1, 1) #plot 1
plt.stem(k, x_k, label="x(k)")
plt.xlabel("k")
plt.ylabel("x[k]")
plt.legend()

plt.subplot(3, 1, 2) #plot 2
plt.stem(k, module_X, label="|X(f)|")
plt.xlabel("Frequence (k)")
plt.ylabel("Amplitude")
plt.legend()

plt.subplot(3, 1, 3) #plot 3
plt.plot(k, argument_X, label="arg(X(f))")
plt.xlabel("Frequence(k)")
plt.ylabel("Phase")
plt.legend()
```
