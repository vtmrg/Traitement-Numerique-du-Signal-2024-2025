III. Phénomène de distribution d'énergie : troncature temporelle, phénomène de Gibss.

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.fft import fft, ifft, fftshift
```

```python
#N=8 
N=8; fo=1/5 # T=1/f=5
Fe=1; Te=1/Fe
k=np.arange(0,N,1)
x_k=np.sin(2*np.pi*fo*k)
t=Te*k


#% on crée un axe des temps 'continu'
# tc=Te*np.arange(0,N,Te)
# xc=np.sin(2*np.pi*fo*tc)

plt.figure()
plt.stem(t,x_k,label='Signal échantilonné') 
# plt.plot(tc,xc,'r',label='Signal "continu"')
plt.xlabel('Temps (ms)');plt.ylabel('Signal')
plt.legend()

pontos_fft=len(x_k)

X_1 = fftshift(fft(x_k, pontos_fft))
x_1 = np.arange(-1/2, 1/2, 1/pontos_fft)
plt.figure()
plt.stem(x_1, np.abs(X_1)/N)
```

```python

pontos_fft=64

X_1 = fftshift(fft(x_k, pontos_fft))
x_1 = np.arange(-1/2, 1/2, 1/pontos_fft)
plt.figure()
plt.stem(x_1, np.abs(X_1)/N)
```
